# 2. Preprocessing of the fruits and vegetables consumption in Spain dataset


In this section, we will perform the corresponding preprocessing of the dataset related to the fruit and vegetables consumption in Spain.

## 2.1. Keeping the selected rows and columns

As mentioned in the data selection part, what we need to do first is to remove the unnecesary rows and columns for our purpose. Lets start with the columns that we will not work with:

```python
initial_consumption_dataset.drop(columns=['CONSUMO X CAPITA', 'VALOR (Miles Euros)', 'VOLUMEN (Miles kg ó litros)'], inplace = True)
initial_consumption_dataset
```

Now, we can remove the rows which represent a cumulative sum of other products or CCAA, like 'Total Nacional', or 'Total Patatas':

```python
initial_consumption_dataset = initial_consumption_dataset[initial_consumption_dataset['CCAA'] != "Total Nacional"]

rows_with_values_to_delete = ["TOTAL PATATAS", "T.HORTALIZAS FRESCAS", "OTR.HORTALIZAS/VERD.", "VERD./HORT. IV GAMA", "T.FRUTAS FRESCAS", "VERDURAS DE HOJA", "FRUTAS IV GAMA"]
initial_consumption_dataset = initial_consumption_dataset[~initial_consumption_dataset['Producto'].isin(rows_with_values_to_delete)]

initial_consumption_dataset
```

## 2.2. Treatment of null values

In the profiling section, we talked about how some values were nulls, lets look at the columns in which they are present:

```python
initial_consumption_dataset.isnull().any()
```

We can check them:

```python
initial_consumption_dataset[initial_covid_dataset.isnull().any(axis = 1)]
```

Most of these nulls represent zero values, regions in which those products where not sold at all (most likely due to climatic conditions for their cultivation). In this case, we will substitute them with zeros to represent their low sellings (because it would be a mistake to eliminate them, as the data is that it was not sold and the model should take that into account):

```python
initial_consumption_dataset["GASTO X CAPITA"] = initial_consumption_dataset["GASTO X CAPITA"].fillna(0)
initial_consumption_dataset["PRECIO MEDIO kg ó litros"] = initial_consumption_dataset["PRECIO MEDIO kg ó litros"].fillna(0)
initial_consumption_dataset
```

We can make sure it worked properly:

```python
initial_consumption_dataset.isnull().any()
```

## 2.3. Checking the type of the features

We will see the features types first:

```python
initial_consumption_dataset.dtypes
```

For better handling of the data, we would ideally want month to be numerical, for what we will use label encoding to replace categorical values with its correspondent number:

```python
initial_consumption_dataset = initial_consumption_dataset.replace("Enero", 1)
initial_consumption_dataset = initial_consumption_dataset.replace("Febrero", 2)
initial_consumption_dataset = initial_consumption_dataset.replace("Marzo", 3)
initial_consumption_dataset = initial_consumption_dataset.replace("Abril", 4)
initial_consumption_dataset = initial_consumption_dataset.replace("Mayo", 5)
initial_consumption_dataset = initial_consumption_dataset.replace("Junio", 6)
initial_consumption_dataset = initial_consumption_dataset.replace("Julio", 7)
initial_consumption_dataset = initial_consumption_dataset.replace("Agosto", 8)
initial_consumption_dataset = initial_consumption_dataset.replace("Septiembre", 9)
initial_consumption_dataset = initial_consumption_dataset.replace("Octubre", 10)
initial_consumption_dataset = initial_consumption_dataset.replace("Noviembre", 11)
initial_consumption_dataset = initial_consumption_dataset.replace("Diciembre", 12)

initial_consumption_dataset['Mes'] = initial_consumption_dataset['Mes'].astype('int64')
initial_consumption_dataset
```

## 2.4. Removing outliers

It is also important to remove the noise-causing outliers. First, we will print the values in the selected columns to confirm that there are some outliers which can be detected with the naked eye:

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (20, 10))

initial_consumption_dataset['GASTO X CAPITA'].plot(alpha = .25, color = 'blue')
```

There are some values which are quite higher than others, one of the even reaching 8, meaning they could be outliers. Lets see the penetration values:

```python
plt.figure(figsize = (20, 10))

initial_consumption_dataset['PENETRACION (%)'].plot(alpha = .25, color = 'blue')
```

In this case, it is not that clear. There are some values which could be outliers. Lets see the average prices now:

```python
plt.figure(figsize = (20, 10))

initial_consumption_dataset['PRECIO MEDIO kg ó litros'].plot(alpha = .25, color = 'blue')
```

It is by far the most notorious case, having values up to 40, while most of them barely surpass 5. But, how can we remove the outliers?

As we will do in the other dataset as well, we will apply the [68-95-99.7 rule](https://en.wikipedia.org/wiki/68%E2%80%9395%E2%80%9399.7_rule), to remove values which are greater than the mean by more than 2/3 times the standard deviation.

```python
consumption_dataset_no_outliers = initial_consumption_dataset.drop(
    initial_consumption_dataset.index[initial_consumption_dataset['GASTO X CAPITA'] > initial_consumption_dataset['GASTO X CAPITA'].mean() + 2 * initial_consumption_dataset['GASTO X CAPITA'].std()]
)

consumption_dataset_no_outliers = initial_consumption_dataset.drop(
    initial_consumption_dataset.index[initial_consumption_dataset['PENETRACION (%)'] > initial_consumption_dataset['PENETRACION (%)'].mean() + 2 * initial_consumption_dataset['PENETRACION (%)'].std()]
)

consumption_dataset_no_outliers = initial_consumption_dataset.drop(
    initial_consumption_dataset.index[initial_consumption_dataset['PRECIO MEDIO kg ó litros'] > initial_consumption_dataset['PRECIO MEDIO kg ó litros'].mean() + 2 * initial_consumption_dataset['PRECIO MEDIO kg ó litros'].std()]
)
```

We will print the dataset with and without outliers for a propper comparison:

```python
plt.figure(figsize = (20, 10))

initial_consumption_dataset['GASTO X CAPITA'].plot(alpha = .25, color = 'blue')
consumption_dataset_no_outliers['GASTO X CAPITA'].plot(alpha = .25, color = 'red')
```

Here, no outliers are removed at all, probably due to the desviation being quite high. This way, higher values will not have a huge impact in our case, so there is not any problem with them. Lets check the next one:

```python
plt.figure(figsize = (20, 10))

initial_consumption_dataset['PENETRACION (%)'].plot(alpha = .25, color = 'blue')
consumption_dataset_no_outliers['PENETRACION (%)'].plot(alpha = .25, color = 'red')
```

As we already thought, no outliers are detected in this column as well. Lets see the final column:

```python
plt.figure(figsize = (20, 10))

initial_consumption_dataset['PRECIO MEDIO kg ó litros'].plot(alpha = .25, color = 'blue')
consumption_dataset_no_outliers['PRECIO MEDIO kg ó litros'].plot(alpha = .25, color = 'red')
```

In this last case, a lot of outliers were removed, not only the most notable ones, but another group of high values.

## 3.6. Removing outliers

Additionally, we also have to perform the outlier detection in order to give some treatment to outliers. However, the question is: how do we detect outliers? The question simply happens because having more cases of covid in 1 day does not necessarily mean that day is an outlier, in fact, it can just be due to the fact that there is a greater population in that territory.

In order to overcome this problem, we decided to create another column for our dataset, "cases_index", which is an index resulting of the division of the cases in one day and the population of the territory where those cases were detected. Thus, with this new column we would be able to compare the different records to detect and treat outliers:

```python
initial_covid_dataset['cases_index'] = initial_covid_dataset.apply(lambda row: row.cases / row.popData2019, axis = 1)
```

Following the same reasoning, we have to do the exact same but with the deaths, creating a new column called 'deaths_index':

```python
initial_covid_dataset['deaths_index'] = initial_covid_dataset.apply(lambda row: row.deaths / row.popData2019, axis = 1)
```

Now, we can simply calculate the descriptive statistics for these new columns, so as to gain understanding of them:

```python
initial_covid_dataset['cases_index'].describe()
```

```python
initial_covid_dataset['deaths_index'].describe()
```

Just by taking a look at these data, we can see how there must be some outliers. If we take a look at the maximum value for the cases_index column, it is ~0.036, which means that in one single day 3.6% of a population got infected, which is obviously an outlier and must be removed (this probably happened in a really small population, and it is generating noise). Same logic can be applied to the maximum value of the death_index column, as it is much higher than the mean value.

As a result, we will be interested in detecting outliers for 3 different features of the dataset:

1. popData2019: because really small or big countries could create noise in the data.

2. cases_index: in order to consider outliers of cases per population (because bigger populations will lead to more cases, so we need to study cases per population, not only cases).

3. deaths_index: in order to consider outliers of deaths per population (because bigger populations will lead to more deaths, so we need to study deaths per population, not only deaths by itself).

Therefore, what we will do is:

- Detect countries that could be outliers (and treat those outliers, probably removing them to avoid noise).
- Detect data records of countries (countries that are not outliers) that could be outliers (essentially, the idea is to detect values that could be outliers within countries whose population is not considered outlier). This is done for both cases_index and deaths_index columns.

### 3.6.1. Outlier detection with 68-95-99.7 rule

There are multiple ways of detecting outliers (we could even apply machine learning techniques such as DBSCAN for outlier detection). However, we decided to apply a more mathematical approach, by making use of the [68-95-99.7 rule](https://en.wikipedia.org/wiki/68%E2%80%9395%E2%80%9399.7_rule), which is based on removing those values that are greater than the mean by more than 2/3 of the standard deviation. Let us apply such technique to our case study:

```python
covid_dataset_no_outliers = initial_covid_dataset.drop(
    initial_covid_dataset.index[initial_covid_dataset['popData2019'] > initial_covid_dataset['popData2019'].mean() + 2 * initial_covid_dataset['popData2019'].std()]
)

covid_dataset_no_outliers = covid_dataset_no_outliers.drop(
    covid_dataset_no_outliers.index[covid_dataset_no_outliers['cases_index'] > covid_dataset_no_outliers['cases_index'].mean() + 2 * covid_dataset_no_outliers['cases_index'].std()]
)

covid_dataset_no_outliers = covid_dataset_no_outliers.drop(
    covid_dataset_no_outliers.index[covid_dataset_no_outliers['deaths_index'] > covid_dataset_no_outliers['deaths_index'].mean() + 2 * covid_dataset_no_outliers['deaths_index'].std()]
)
```

The following dataset is the one with no outliers according to this technique:

```python
covid_dataset_no_outliers
```

We can visualize the results to see which outliers were removed. For cases index:

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (20, 10))

initial_covid_dataset['cases_index'].plot(alpha = .25, color = 'blue')
covid_dataset_no_outliers['cases_index'].plot(alpha = .25, color = 'red')
```

As we can see from the picture, among severeal outliers that were removed, we did remove the day where 3.6% of the population got infected by covid-19 (as it was commented earlier, this probably is because of a small population getting several members infected).

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (20, 10))

initial_covid_dataset['deaths_index'].plot(alpha = .25, color = 'blue')
covid_dataset_no_outliers['deaths_index'].plot(alpha = .25, color = 'red')
```

As we can see from the picture, among several outliers that were removed, we did remove the day where the highest deaths occurred (which was ~1000x higher than the average) in a population.

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (20, 10))

initial_covid_dataset['popData2019'].plot(alpha = .25, color = 'blue')
covid_dataset_no_outliers['popData2019'].plot(alpha = .25, color = 'red')
```

As we can see, there are some countries that have really high values for popData2019, we can even try to identify these countries:

```python
countries_high_population = initial_covid_dataset[initial_covid_dataset['popData2019'] > 10**9]
countries_high_population['countriesAndTerritories'].unique()
```

As we can see, both China and India are the countries that are detected as outliers, whereas the rest of the values are not considered as outliers by this statistical technique.

Additionally, we can see how the treatment that we should give to the outliers could essentially be:

- Replace with an average value.
- Remove the outlier.

In this case, we strongly believe that, having a dataset as big as we have, the best treatment could be to remove these values:

- For the outliers related to the columns cases_index and deaths_index, values that are too high could be mistaken, due to a count error (counting on a day, data records from different days), it is true that we could do some sort of average of the values, but we do not know the real time interval of those cases (you can create an average, but that would lead to introduced values that are made up, which could also create noise in the model).

- For the outliers related to popData2019, a decision must be performed to either keep China and India, or to not keep them. Since these countries have a population that is much higher than other populations (of the order of 5-6 times higher than already big populations) we do consider these countries as outliers, thus we will remove the values associated to these two countries, keeping the rest of them.

From this point on, we will use this new dataset `covid_dataset_new_outliers` will be used for the data preprocessing, as it does not contain those outliers.



### 3.6.2. Anomaly detection with BigML

In order to achieve the best results possible, we considered the use of BigML in order to perform a double check related to the outliers detection (that was previously performed with a statistical rule). 

Thus, the idea is to use BigML in order to see general data related to the data cleaning (to see how good or bad we cleaned the data) and also we want to identify possible outliers that were undetected with the previous technique.

The first step is to download the covid-19 dataset after applying all the operations that we performed on it at this moment, so that we can use such dataset as the initial dataset for the analysis with BigML. 

```python
covid_dataset_no_outliers.to_csv('dataset5_before_bigML.txt')
```

Once uploaded the datafile as a source for BigML, we can see how BigML gives us 2 warnings related to the features:

feature_warnings.png

- Warning on "year" column: the first warning is related to the fact that the "year" column only contains 1 value. For our data card, as it will be mentioned later on, this field will not be taken into account as this problem was already known.

- Warning on "countriesAndTerritories": the second warning refers to the fact that there are lots of values for that categorical and nominal variable. However, this again is not that much of a problem, because we will end up aggregating the data to compute the data with which we will enrich the consumption dataset. Thus, it is again a problem that will not happen in our final data cards.

We can also notice, just by taking a look at the the 'cases_index' and 'death_index' features (which are the ones that take into account cases/deaths per population) how the distribution is clearly right skewed, so there will probably still be some outliers in our data. Thus, we have to perform the anomaly detection in order to detect possible outliers.

In order to do that, we created an anomaly detector in BigML, that used the covid-19 dataset as it currently is (after removing some of the outliers with the previous statistical technique). Afterwards, we computed a new column using BigML called "score" which represents the anomaly score of each data record (it represents from 0 to 1 how much of an outlier they are, where 0 is no outlier and 1 is complete outlier). According to [BigML's documentation](https://bigml.com/features/anomaly-detection), values of 60% or higher typically represent outliers. As a result, we will be deleting those values. But first, let us import the dataset with this new column obtained from BigML: 

```python
covid_dataset_after_bigML = pd.read_csv('/content/dataset5_after_bigML.txt')
covid_dataset_after_bigML['score'].describe()
```

As we can see from the descriptive stastistics, the mean of the values is about 0.36, which is a really good value.

However, if we look at the maximum value, it is about 0.75, which means that there are still some outliers that were not detected by the previous technique.

Let's show the potential outliers, using 0.6 as the value from which a data record will be considered outlier (again, this comes from [BigML's](https://bigml.com/features/anomaly-detection) documentation):

```python
covid_dataset_after_bigML[covid_dataset_after_bigML['score'] >= 0.6]
```

As we can see, there are 1200 records whose anomaly score, computed by BigML, is higher or equal than 0.6. These 1200 records are considered outliers, and, as a result, we should give them a treatment. 

In this case, as well as we did in the previous techique, we consider that the best treatment for the outliers is to remove them. This is because there could be multiple reasons why they are outliers:

- Maybe the cases of multiple days were aggregated and showed in a data record (with a higher value of covid cases/deaths).
- Maybe the population is so small that having a small amount of covid cases, the columns cases_index and deaths_index have really high values in comparison to countries with a higher population
- Maybe it is due to a human mistake.

As a result, since it is not possible to know the exact reason, we consider that doing an average would be creating noise (as we create values that might not be the real ones and might not even be similar at all), so we decided to remove those values. Thus, the covid-19 dataset after dealing with outliers is the following one: 

```python
bigML_outliers = covid_dataset_after_bigML[covid_dataset_after_bigML['score'] >= 0.6]
covid_dataset_no_outliers.drop(bigML_outliers['field1'], inplace = True)
covid_dataset_no_outliers
```
