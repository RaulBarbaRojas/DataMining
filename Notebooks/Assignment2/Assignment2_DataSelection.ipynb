## 2.1. Data selection for dataset 1

Form that dataset we will exclude certain columns, because they are not useful for pur purposes. The columns that we are not keeping are the following ones:
- "Volumen (miles de kg)": this column describes the amount (in thousands of kg) that were sold for a given product (given the month and the year). However, for our purposes, we strongly believe that it will not be useful (we would need a feature that describes how optimal a product is based not only in amount or only in price, but based on the expense per person in the product).
- "Valor (miles de €)": it is a column that is associated to the previous one. It describes the amount of money (in thousands of euros) that the volume of the product costed. Again, we remove it for the same reason as before. 
- "Consumo x capita": the main reason why the column is deleted is because it does not give valuable information regarding how optimal a product is. In the end, a product could have a really high consumption per capita, but such a low price that it could be a not so optimal product (for us to plant and sell). Thus, the column is not valuable for our purposes and it should be deleted.

```python
dataset_1.drop(columns=['VOLUMEN (Miles kg ó litros)', 'VALOR (Miles Euros)', 'CONSUMO X CAPITA'], inplace = True)
```

Moreover, we must consider that the column "PENETRACION (%)" is only useful for some of our purposes, but not for all of them (in the data preprocessing we deal with this column with more detail, for now it is simply included).

We will also delete each row that contains value "Total Nacional" in column CCAA, because these is simply an aggregation of the different values of each region. As a result, it is not "new data", but simply data that we already have. Thus, for our purposes we will perform all the operations based on the real data.

```python
dataset_1 = dataset_1[dataset_1['CCAA'] != "Total Nacional"]
dataset_1
```

Also for similar reason we exclude rows that in column "Producto" that have name that indicates this row may contain sum of other values. Those names are: "TOTAL PATATAS", "T.HORTALIZAS FRESCAS", "OTR.HORTALIZAS/VERD", "VERD./HORT. IV GAMA", "T.FRUTAS FRESCAS", "VERDURAS DE HOJA", "FRUTAS IV GAMA". It gives us assurance that none of the products is counted twice which could lead to false conclusions. Next thing is that one of our objectives is to predict which product to chose and predicting for example frutas IV gama would be too general.

```python
rows_with_values_to_delete = ["TOTAL PATATAS", "T.HORTALIZAS FRESCAS", "OTR.HORTALIZAS/VERD.", "VERD./HORT. IV GAMA", "T.FRUTAS FRESCAS", "VERDURAS DE HOJA", "FRUTAS IV GAMA"]
dataset_1 = dataset_1[~dataset_1['Producto'].isin(rows_with_values_to_delete)]
```

```python
dataset_1.sample(10)
```
