Why could these data be null? Well, by definition, the incidence is measured according to the cumulative covid cases in the past 14 days of a country. Thus, the first 13 records for each country will not provide an incidence, because we did not cumulate 14 days. However, we need to check whether this theory is true or not. The data that we just obtained is really interesting, because technically, for each each country there should be 13 data records that we should remove (because the cumulative incidence does not make sense for cases in which we do not have the 14 values). ### 2.5. Outliers

Within the exploratory data analysis, another important part is to try to determine whether we have outliers or not. We will apply statistical outlier detection techniques.

Specifically, we can apply the rule called: [68-95-99,7](https://es.wikipedia.org/wiki/Regla_68-95-99.7), also known as "empirical rule", which is based on considering as outliers any data whose value is higher than the mean by more than 2/3 of the standard deviation. It can be applied in a simple way:

```python
no_outliers_covid_dataset = covid_dataset.drop(covid_dataset.index[
    (covid_dataset['Cumulative_number_for_14_days_of_COVID-19_cases_per_100000'] > (covid_dataset['Cumulative_number_for_14_days_of_COVID-19_cases_per_100000'].mean() + 2 * covid_dataset['Cumulative_number_for_14_days_of_COVID-19_cases_per_100000'].std()))
])

no_outliers_covid_dataset
```

```python
print(f'Number of detected outliers: {len(covid_dataset) - len(no_outliers_covid_dataset)}')
```
